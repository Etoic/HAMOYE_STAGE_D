```python
#Initialize the Total sum of Food produced(Kiloton) and Food Supplied(Kcal/person day) by Year
df3 = df1.groupby('Year', as_index=False).Value.sum()
df4 = df2.groupby('Year', as_index=False).Value.sum()
```

```python
#See the trend of Food produced(in Kiloton) per year. 
df3.iplot(kind='line',x='Year',y='Value',title='FOOD PRODUCTION TREND BY YEAR',xTitle='Year',
          yTitle='Food Produced in Kiloton')
```

**Apart for the drop between 2006 and 2007, there was a progressive increase in food production**

```python
df4.iplot(kind='line',x='Year',y='Value',title='FOOD SUPPLY TREND BY YEAR',xTitle='Year',
          yTitle='Food Supplied in Kcal/(Person Day)')
```

**Progressive increase in food supplied**

```python
#Initialize the Total sum of Food produced(Kiloton) and Food Supplied(Kcal/person day) by Country
df5 = df1.groupby('Country', as_index=False).Value.sum()
df6 = df2.groupby('Country', as_index=False).Value.sum()
```

```python
#See the Volume of Food produced per Country.
fig = px.bar(df5, x="Country", y="Value",
                 height=1000, labels={'Item':'Food Items', 'Value':'Volume of Food Produced'' (Kiloton)'},
            title='VOLUME OF FOOD PRODUCED BY AFRICAN COUNTRIES'' (2004-2013)')
fig.show()
```

**Top 3 Producing Countries within the period under review: Nigeria, Egypt, & South Africa.**

```python
#See the Volume of Food Supplied per Country.
fig = px.bar(df6, x="Country", y="Value",
                 height=1000, labels={'Item':'Food Items', 'Value':'Volume of Food Supplied'' (Kcal/Person Day)'},
            title='VOLUME OF FOOD SUPPLIED BY AFRICAN COUNTRIES'' (2004-2013)')
fig.show()
```

**Top 3 Supplying Countries within the period under review: Egypy, Tunisia, & Morocco.**

```python
#Initialize the Total sum of Food produced(Kiloton) by Food Items
df7 = df1.groupby('Item', as_index=False).Value.sum()

fig = px.bar(df7, x="Item", y="Value",
                 height=1000, labels={'Item':'Food Items', 'Value':'Volume Produced (Kilotons)'},
            title = 'VOLUME OF FOOD PRODUCED BY FOOD ITEMS (2004-2013)')
fig.show()
```

**Cassava, Sugar Cane, and Maize are the top three food items produced within the period under review.**

```python
#Let's put the spotlight on the Top 3 producing countries.
Top_1 = df1[df1['Country'] == 'Nigeria']
Top_2 = df1[df1['Country'] == 'Egypt']
Top_3 = df1[df1['Country'] == 'South Africa']
```

```python
Top1_food = Top_1.groupby(['Item'], as_index=False).Value.sum()
Top2_food = Top_2.groupby(['Item'], as_index=False).Value.sum()
Top3_food = Top_3.groupby(['Item'], as_index=False).Value.sum()
```

```python
fig = px.bar(Top1_food, x="Item", y="Value",
                 height=1000, labels={'Item':'Food Items', 'Value':'Volume Produced (Kilotons)'},
            title = 'VOLUME OF FOOD PRODUCED BY NIGERIA (2004-2013)')
fig.show()
```

**Nigeria's highest produces are Cassava(and its products) & Yams**

**Cassava and Yam are similar in their Tuberous nature, which suggests the relative ease in the production of Tuberous products in Nigeria. Furthermore, from my physical observation, anyone with a small piece of land can easily cultivate Cassava in its season. I reckon the sames applies to Yam**

**While Nigeria overwhelming leads the food production scale in Africa, it is heavily reliant on two products, and needs to do better with across other products.**

```python
fig = px.bar(Top2_food, x="Item", y="Value",
                 height=1000, labels={'Item':'Food Items', 'Value':'Volume Produced (Kilotons)'},
            title = 'VOLUME OF FOOD PRODUCED BY EGYPT (2004-2013)')
fig.show()
```

**Egypt has Sugar Cane has its number one produce.**

**Unlike Nigeria, we see a relatively fair spread in food production. Products like Tomatoes, Wheat, Vegetables, and Maize account for the relatively good spread.**

```python
fig = px.bar(Top3_food, x="Item", y="Value",
                 height=1000, labels={'Item':'Food Items', 'Value':'Volume Produced (Kilotons)'},
            title = 'VOLUME OF FOOD PRODUCED BY SOUTH AFRICA (2004-2013)')
fig.show()
```

```python
#A few statistical plot to represent means, quartiles, medians, etc.

#Mean values of Food Produced per Year
df1_mean = df1.groupby('Year', as_index=False).Value.mean()
df1_mean
```

```python
#Mean values of Food Supplied per Year
df2_mean = df2.groupby('Year', as_index=False).Value.mean()
df2_mean
```

```python
#A box plot for Food Produced by Year
box1 = px.box(df1, x='Year', y='Value', hover_name='Country', labels={'Year':'Year', 'Value':'Value'},
                title = 'A BOX PLOT FOR FOOD PRODUCED (2004-2013)')
box1
```

```python
#A box plot for Food Produced by Year
box2 = px.box(df2, x='Year', y='Value', hover_name='Country', labels={'Year':'Year', 'Value':'Value'},
                title = 'A BOX PLOT FOR FOOD SUPPLIED (2004-2013)')
box2
```

```python
#A side by side look at the trend of average Food Production & Supply per year.
fig = make_subplots(rows=2, cols=1,
                    shared_xaxes=True,
                    vertical_spacing=0.02)

fig.add_trace(go.Line(x=df1_mean['Year'], y=df1_mean['Value'], name='Mean Food Production'),
              row=2, col=1)

fig.add_trace(go.Line(x=df2_mean['Year'], y=df2_mean['Value'], name='Mean Food Supply'),
              row=1, col=1)

fig.update_layout(height=1000, width=1000,
                  title_text="Mean Food Production & Supply between 2004-2013")

fig.show()
```
